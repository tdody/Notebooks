*import libraries*

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
```

# Histograms

```python
data = np.random.normal(size=1000)

fig,axes = plt.subplots(figsize=(5,5))
hist = axes.hist(data,bins=50,alpha=0.5)
```

# Pies

```python
fracs = [30, 15, 45, 10]

fig,axes = plt.subplots(figsize=(5,5))
pie = axes.pie(fracs,labels=['A','B','C','D'],explode=[0.08,0,0,0])
```

# Errorbar

```python
x = np.linspace(0,10,40)
dy = 0.1 #standard dev
y = np.random.normal(loc=sin(x),scale=dy)

real = sin(x)

fig,axes = plt.subplots(figsize=(15,8))
axes.errorbar(x,y,dy,fmt='.k')
axes.plot(x,real,color='red',linestyle=':')

axes.spines['left'].set_position('zero')
axes.spines['right'].set_color('none')
axes.spines['bottom'].set_position('zero')
axes.spines['top'].set_color('none')
axes.set_xlim(0,10)
```
