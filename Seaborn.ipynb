# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math

%matplotlib inline
```

```python
#Seaborn settings
sns.set_palette("deep",desat=0.6)
sns.set_context(rc={"figure.figsize":(8,4)})
sns.set_style("whitegrid")
```

## Import data

```python
titanic = pd.read_csv('./Seaborn Data/titanic.csv')
```

```python
server = pd.read_csv('./Seaborn Data/serverdata.csv')
```

```python
mlb = pd.read_csv('./Seaborn Data/mlbsalaries.csv')
```

## Plots

## Histograms

```python
plt.hist(titanic.Age.dropna(), bins=25,edgecolor="k");
plt.title("Passenger age distribution")
plt.xlabel("Age")
plt.ylabel("Count");
```

```python
plt.hist(server.Time,bins=25,edgecolor="k",color=sns.desaturate("indianred",1));
plt.title("Server response time")
plt.xlabel("Response [ms]")
plt.ylabel("Count");
```

```python
plt.hist(mlb.Salary.dropna(), bins=25,edgecolor='k',color=sns.desaturate("orange",1));
plt.xlim(0,max(mlb.Salary.dropna()))
plt.title("MLB salary distribution")
plt.xlabel("Salary [US$]")
plt.ylabel("Count");
```

## Boxplot

```python
sns.boxplot(titanic.Age.dropna(),color="green")
plt.title("Titanic age distribution")
```

Group boxplots:

```python
sns.boxplot(titanic.Age, titanic.Sex);
```

```python
sns.boxplot(server.Time);
```

```python
sns.boxplot(mlb.Salary);
```

## Kernel Density Estimate Plot

```python
sns.kdeplot(titanic.Age.dropna(), shade=True,color=sns.desaturate("plum",1));
```

```python
sns.kdeplot(server.Time.dropna(), shade=True,color=sns.desaturate("indianred",1));
```

```python
sns.kdeplot(mlb.Salary.dropna(), shade=True,color=sns.desaturate("mediumaquamarine",1));
```

```python
sns.distplot(titanic.Age.dropna(),color=sns.desaturate("plum",1));
```

```python
sns.distplot(server.Time.dropna(),color=sns.desaturate("indianred",1));
```

```python
sns.distplot(mlb.Salary.dropna(), color=sns.desaturate("mediumaquamarine",1));
```

## Violin plot

```python
sns.violinplot(titanic.Age.dropna());
```

```python
sns.violinplot(titanic.Age.dropna(), titanic.Sex,palette="muted", inner="quartile");
```

```python
sns.violinplot(mlb.Salary,palette="Set2");
```

## Cumulative distribution function

```python
sns.kdeplot(titanic.Age.dropna(), cumulative=True,shade=True);
```

```python
sns.kdeplot(server.Time.dropna(), cumulative=True,shade=True,
            color=sns.desaturate("yellow",0.7));
```

```python
sns.kdeplot(mlb.Salary.dropna(), cumulative=True,shade=True);
```
