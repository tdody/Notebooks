## 2. matplotlib object-oriented API

The main idea behing the object-oriented plotting is to apply functions and actions on the plot object.

### 2.1. General

```python
# Instantiate the figure.
fig = plt.figure()

# Generate content of the plot
axes = fig.add_axes([0,0,1,1]) # Add axes at position [left, bottom, width, height]

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('f(x)=x^2')
axes.set_xlim(0,5)
axes.set_ylim(0,25)
```

```python
fig = plt.figure()
axes_main = fig.add_axes([0,0,1,1])
axes_sub = fig.add_axes([0.2,0.5,0.4,0.4])

# Main plot.
axes_main.plot(x,y,'r')
axes_main.set_xlabel('x')
axes_main.set_ylabel('f(x)')
axes_main.set_title('f(x)=x^5')
axes_main.set_xlim(0,5)
axes_main.set_ylim(0,25)

# Sub plot.
axes_sub.plot(x,y,'g--')
axes_sub.set_xlabel('x')
axes_sub.set_ylabel('f(x)')
axes_sub.set_title('f(x)=x^0.5')
axes_sub.set_xlim(0,5)
axes_sub.set_ylim(0,25);
```

Another fast way to create tthe figure and add a default axes object to it is to use subplots(). This function returns a figure and an axes obects.

```python
fig, axes =plt.subplots()

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('f(x)=x^2')
axes.set_xlim(0,5)
axes.set_ylim(0,25);
```

```python
# Multiplot ~ subplots
fig, axes = plt.subplots(nrows=1,ncols=2)

for ax in axes:
    ax.plot(x,y,'r')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
```

### 2.2. Figure size, aspect ratio and DPI

Matplotlib can be used to set:
- the aspect ratio
- DPI
- figure size 
Using the figsize and dpi keyword arguments. figsize is a tuple of the width and height of the figure in inches, and dpi is the dots-per-inch (pixel per inch). To create an 800x400 pixel, 100 dots-per-inch figure, we use figure(figsize(8,4),dpi=100).

```python
fig, axes = plt.subplots(figsize=(6,3),dpi=100)
axes.plot(x, x**2, 'r',label = "f(x)=x^2")
axes.plot(x, x**1.5, 'b',label = "f(x)=x^1.5")
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
axes.legend();
```

## 2.5. Colors, linewidths, linetypes

- 'b.-' = blue line with dots
- 'g--' = green line with dashes

The line style can be specified using ls=, the color by using color=

```python
fig, ax = plt.subplots()

ax.plot(x, x+1,ls='-',color="red")
ax.plot(x, x+2,ls='--',color="green")
ax.plot(x, x+3,ls=':',color="blue");
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x**2+1, color="blue", linewidth=0.25)
ax.plot(x, x**2+2, color="blue", linewidth=0.50)
ax.plot(x, x**2+3, color="blue", linewidth=1.00)
ax.plot(x, x**2+4, color="blue", linewidth=2.00)

# Possible linestype options -, --, -., :, steps
ax.plot(x, x**2+5, color="red", lw=2, linestyle='-')
ax.plot(x, x**2+6, color="red", lw=2, ls='-.')
ax.plot(x, x**2+7, color="red", lw=2, ls=':')

# Custom dash
line, = ax.plot(x, x**2+8, color="black", lw=1.50)
line.set_dashes([10, 5, 20, 5]) # format: line length, space length, ...

# Possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x**2+ 9, color="green", lw=2, ls='--', marker='+')
ax.plot(x, x**2+10, color="green", lw=2, ls='--', marker='o')
ax.plot(x, x**2+11, color="green", lw=2, ls='--', marker='s')
ax.plot(x, x**2+12, color="green", lw=2, ls='--', marker='1')

# Marker size and color
ax.plot(x, x**2+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x**2+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x**2+15, color="purple", lw=1, ls='-', marker='o', markersize=8, 
        markerfacecolor="red")
ax.plot(x, x**2+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=2, markeredgecolor="blue");
```

### 2.6. Axis appearance

### 2.6.a. Plot range

```python
fig,axes = plt.subplots(1,3,figsize=(12,4))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('Default appearance')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,0.5*max(x**2)])
axes[2].set_xlim([0,0.5*max(x)])
axes[2].set_title('custom range');
```

###  2.6.b. Scale

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")
axes[0].legend([r'$x^2$',r'$exp(x)$'])

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_xscale("log")
axes[1].set_title("Logarithmic scale (y)")
axes[1].legend([r'$x^2$',r'$exp(x)$']);
```

### 2.6.c. Ticks and tick labels

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$',
                    r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18);
# use LaTeX formatted labels
```

### 2.6.d. Scientific notation

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("Scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```

### 2.7. Axis grid

```python
fig,axes = plt.subplots(1,2,figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='g', alpha=0.5, linestyle='dotted', linewidth=0.5)
```

### 2.8. Mutiple axes

```python
fig,ax1 = plt.subplots()

ax1.plot(x,x**2,lw=2,color='blue')
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")

ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")

for label in ax1.get_yticklabels():
    label.set_color('blue')
for label in ax2.get_yticklabels():
    label.set_color('red')
```

**Colormap and Contours**

```python
# Define function to be plotted
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
```

```python
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)

X,Y = np.meshgrid(phi_p, phi_m) # Returns a list of arrays containing the values of p and m
Z = flux_qubit_potential(X, Y).T # Needs to be transposed so X is the abscissa
```

```python
fig, ax = plt.subplots()

p = ax.pcolor(X, Y, Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())
cb = fig.colorbar(p, ax=ax)
```

**imshow**

```python
fig, ax = plt.subplots()

im = ax.imshow(Z, cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), 
               vmax=abs(Z).max(), extent=[0, 2*np.pi, 0, 2*np.pi])
im.set_interpolation('bilinear')

cb = fig.colorbar(im, ax=ax)
```

**countour**

```python
fig,ax = plt.subplots()
contour = ax.contour(Z,cmap=matplotlib.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(),extent=[0, 2*np.pi, 0, 2*np.pi])
```

## 4. 3D Plots

To use 3D graphics in matplotlib, we first need to create an instance of the Axes3D class. 3D axes can be added to a matplotlib figure canvas in exactly the same way as 2D axes; or, more conveniently, by passing a projection='3d' keyword argument to the add_axes or add_subplot methods.

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

**Surface plots**

```python
fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1, 2, 1, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

ax = fig.add_subplot(1, 2, 2, projection='3d')
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=matplotlib.cm.coolwarm, linewidth=0, antialiased=False)
cb = fig.colorbar(p, shrink=0.5)
```

**Wire-frame plot**

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(1,1,1,projection='3d')
p = ax.plot_wireframe(X,Y,Z, rstride=4, cstride=4)
```

**Contour with projection on three plans**

```python
fig = plt.figure(figsize=(8,6))

ax = fig.add_subplot(1,1,1, projection='3d')

ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
cset = ax.contour(X, Y, Z, zdir='z', offset=-np.pi, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='x', offset=-np.pi, cmap=matplotlib.cm.coolwarm)
cset = ax.contour(X, Y, Z, zdir='y', offset=3*np.pi, cmap=matplotlib.cm.coolwarm)

ax.set_xlim3d(-np.pi, 2*np.pi);
ax.set_ylim3d(0, 3*np.pi);
ax.set_zlim3d(-np.pi, 2*np.pi);
```

** Rotate default view**

We can change the perspective of a 3D plot using the view_init method, which takes two arguments: elevation and azimuth angle (in degrees):

```python
fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(30, 45)

ax = fig.add_subplot(1,2,2,projection='3d')
ax.plot_surface(X, Y, Z, rstride=4, cstride=4, alpha=0.25)
ax.view_init(60, 30)

fig.tight_layout()
```
